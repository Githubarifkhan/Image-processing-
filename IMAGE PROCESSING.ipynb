```python
def  title_im(img,n,m):
    t_im_1st=[]
    
    for i in range(n):# n in species  rows
        t_im_1st.append(img)
        
    t_im =np.concatenate(t_im_1st , axis = 1)
    
    t_im_1st = []
    for i in range(m):
         t_im_1st.append(t_im)
            
    t_im =np.concatenate(t_im_1st , axis = 0)
    return t_im
plt.imshow(title_im(im,2,2))
```
